```python
%%writefile projeto_drone.py
print("--- INSTALANDO ULTRALYTICS YOLOv8 ---")
import os
os.system("pip install ultralytics")

from ultralytics import YOLO
from PIL import Image

# 1. Gerar o arquivo de configuração
config_dataset = """
path: ./dataset
train: train/images
val: valid/images
nc: 2
names: ['Lixo_Plastico', 'Mancha_Oleo']
"""
with open("data.yaml", "w") as f:
  f.write(config_dataset)
print("Arquivo 'data.yaml' gerado!")

# 2. Criar uma imagem de teste local (sem usar a internet)
print("\n--- CRIANDO IMAGEM DE TESTE LOCAL ---")
img = Image.new('RGB', (640, 640), color = 'blue')
img.save("teste_drone.jpg")
print("Imagem 'teste_drone.jpg' criada localmente!")

# 3. Carregar o modelo
print("\n--- CARREGANDO MODELO NEURAL NANO ---")
model = YOLO("yolov8n.pt")

# 4. Rodar a Inteligência Artificial na imagem criada
print("\n--- DRONE EM OPERAÇÃO: ANALISANDO IMAGEM ---")
resultados = model.predict(source="teste_drone.jpg", save=True, conf=0.25)

print("\n--- DIAGNOSTICO DA VISAO COMPUTACIONAL ---")
print("SUCESSO: A IA processou a imagem de teste!")
```
